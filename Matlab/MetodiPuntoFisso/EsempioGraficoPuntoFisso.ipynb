# Esercitazione punto fisso

### Esercizio 1: usare iterazioni di punto fisso

Calcolare il punto fisso di:<br>
$x_k+1 = \phi(x) = cos(x_k)$<br>
calcolate il decimo elemento quando:
<ul>
    <li>$x_0 = 1$</li>
    <li>$x_0 = 1.2$</li>
    <li>$x_0 = 0.5$</li>
</ul>

Per ciascu caso disegnate su un grafico dell'andamento degli incrementi.<br>
Ripetete il caso per:<br>
$x_k+1 = \phi(x) = \frac{x^2_k + 2}{2x_k}$

```matlab
%plot funzione di partenza:
f =@(t) cos(t) - t;
dominio = linspace(-5,5);
plot(dominio,f(dominio),dominio,zeros(length(dominio)));
```

Quindi quello che interessa a noi è trovare la x tale per cui cos(x) - x = 0. Per far questo andiamo a creare una funzione phi = cos(x) e calcoliamo il punto di intersezione con la retta y(x) = x i punti di phi e f coincideranno.

```matlab
%Definizione della funzione f(x) = cos(x) - x, ricordiamo che per definizione 
% phi è una funzione del tipo phi(x) = x per cui avendo cos(t) - x -> cos(t) = x quindi
% phi = cos(t)

retta = @(t) t;
phi = @(t) cos(t);

```

```matlab
x0 = 1;
nMax = 10;
n = 0;
INC = [];
%Plot della funzione iniziale cos(x) - x

plot(dominio,phi(dominio),dominio,retta(dominio));
hold on
plot(dominio,zeros(length(dominio),1),'b',zeros(length(dominio),1),linspace(-5,5,length(dominio)),'b');

while(n<nMax)
    x = phi(x0);
    t = linspace(x0,x,100);
    plot(x0*ones(100,1),t,'k');
    plot(linspace(x0,x),x*ones(100,1),'k');
    INC(end + 1) = x - x0;
    x0 = x;    
    n=n+1;
end
```
